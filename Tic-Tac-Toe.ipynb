```python
# TIC-TAC-TOE

def printBoard(xState, zState):
    
    zero = 'X' if xState[0] else ('O' if zState[0] else 0)
    one = 'X' if xState[1] else ('O' if zState[1] else 1)
    two = 'X' if xState[2] else ('O' if zState[2] else 2)
    three = 'X' if xState[3] else ('O' if zState[3] else 3)
    four = 'X' if xState[4] else ('O' if zState[4] else 4)
    five = 'X' if xState[5] else ('O' if zState[5] else 5)
    six = 'X' if xState[6] else ('O' if zState[6] else 6)
    seven = 'X' if xState[7] else ('O' if zState[7] else 7)
    eight = 'X' if xState[8] else ('O' if zState[8] else 8)
    
    print(f" {zero} | {one} | {two} ")
    print(f"---|---|---")
    print(f" {three} | {four} | {five} ")
    print(f"---|---|---")
    print(f" {six} | {seven} | {eight} ") 
    
def printBoardFinal(xState, zState,index1,index2,index3):
    
    if 0 == index1 or 0 == index2 or 0 == index3:
        zero = '*' if xState[0] else ('*' if zState[0] else 0)
    else:
        zero = 'X' if xState[0] else ('O' if zState[0] else 0)
        
    if 1 == index1 or 1 == index2 or 1 == index3:
        one = '*' if xState[1] else ('*' if zState[1] else 1)
    else:
        one = 'X' if xState[1] else ('O' if zState[1] else 1)
    
    if 2 == index1 or 2 == index2 or 2 == index3:
        two = '*' if xState[2] else ('*' if zState[2] else 2)
    else:
        two = 'X' if xState[2] else ('O' if zState[2] else 2)
    
    if 3 == index1 or 3 == index2 or 3 == index3:
        three = '*' if xState[3] else ('*' if zState[3] else 3)
    else:
        three = 'X' if xState[3] else ('O' if zState[3] else 3)
    
    if 4 == index1 or 4 == index2 or 4 == index3:
        four = '*' if xState[4] else ('*' if zState[4] else 4)
    else:
        four = 'X' if xState[4] else ('O' if zState[4] else 4)
        
    if 5 == index1 or 5 == index2 or 5 == index3:
        five = '*' if xState[5] else ('*' if zState[5] else 5)
    else:
        five = 'X' if xState[5] else ('O' if zState[5] else 5)
    
    if 6 == index1 or 6 == index2 or 6 == index3:
        six = '*' if xState[6] else ('*' if zState[6] else 6)
    else:
        six = 'X' if xState[6] else ('O' if zState[6] else 6)
        
    if 7 == index1 or 7 == index2 or 7 == index3:
        seven = '*' if xState[7] else ('*' if zState[7] else 7)
    else:
        seven = 'X' if xState[7] else ('O' if zState[7] else 7)
        
    if 8 == index1 or 8 == index2 or 8 == index3:
        eight = '*' if xState[8] else ('*' if zState[8] else 8)
    else:
        eight = 'X' if xState[8] else ('O' if zState[8] else 8)
        
    print(f" {zero} | {one} | {two} ")
    print(f"---|---|---")
    print(f" {three} | {four} | {five} ")
    print(f"---|---|---")
    print(f" {six} | {seven} | {eight} ") 
    
    

def get_current_state(xState, zState):
    
    wins = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6]]
    for win in wins:
        if(xState[win[0]] + xState[win[1]] + xState[win[2]] == 3):
            print("X Won the match")
            print()
            printBoardFinal(xState, zState,win[0],win[1],win[2])
            print()
            return 1
        if(zState[win[0]] + zState[win[1]] + zState[win[2]] == 3):
            print("O Won the match")
            print()
            printBoardFinal(xState, zState,win[0],win[1],win[2])
            print()
            return 0
    return -1
    

ans = "y"
while(ans == "y" or ans == "Y"):
    
    visited = [False for i in range(9)] # for checking if both X and O enter the value at same place
    count = 0
    xState = [0, 0, 0, 0, 0, 0, 0, 0, 0] # initial state for X before starting the game
    zState = [0, 0, 0, 0, 0, 0, 0, 0, 0] # initial state for O before starting the game
    
    turn = 1 # 1 for X and 0 for O
    
    print("Welcome to Tic Tac Toe")
    print()
    while(True):
        printBoard(xState, zState)
        print()
        if(count == 9):
            print("No one wins, it is a tie :)")
            print("Do you want to play again? --> enter y for yes and press any other key to exit")
            ans = input()
            break
            
        if(turn == 1):
            print("X's Chance")
            print()
            value = int(input("Please enter a value: "))
            print()
            if visited[value] == True:
                print("Wrong Move!!,Match over O won")
                print("Do you want to play again? --> enter y for yes and press any other key to exit")
                ans = input()
                break
            else:
                visited[value] = True
                xState[value] = 1
                count += 1
                
        else:
            print("O's Chance")
            print()
            value = int(input("Please enter a value: "))
            print()
            if visited[value] == True:
                print("Wrong Move!!,Match over X won")
                print("Do you want to play again? --> enter y for yes and press any other key to exit")
                ans = input()
                break
            else:
                visited[value] = True
                zState[value] = 1
                count += 1
                
        checkWin = get_current_state(xState, zState)
        
        if(checkWin != -1):
            print("Match over")
            print("Do you want to play again? --> enter y for yes and press any other key to exit")
            ans = input()
            break
            
        turn = 1 - turn
        
print("Game over")    
    
```
